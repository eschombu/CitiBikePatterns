```python
import numpy as np
import pandas as pd
from datetime import date, datetime, timedelta
import datetime as dt
from pytz import timezone
import pytz
import time
import calendar
# import matplotlib
# matplotlib.use(‘Agg’)
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
def ismember(A, B):
    return [ bool(np.sum(a == B)) for a in A ]
```

```python
# Load Citi Bike data
CBmonths = np.array([np.array([np.ones((6,1))*2013, np.ones((6,1))*2014]).flatten(), \
    np.array([np.arange(7,13,1), np.arange(1,7,1)]).flatten()], dtype='int').T
mo_strs = []
for i in range(CBmonths.shape[0]):
    mo_strs.append(date(CBmonths[i,0], CBmonths[i,1], 1))

citibike_filename = "/home/vagrant/citibike/hourlyDB_" + mo_strs[0].strftime("%m-%Y") + \
    "_to_" + mo_strs[-1].strftime("%m-%Y") + ".csv.gz"
hourlyDF_bystart = pd.read_csv(citibike_filename, compression="gzip")

# Get proper DateTime objects
hourlyDF_bystart["DateTime"] = hourlyDF_bystart["startbin"].apply(lambda x: \
    datetime(CBmonths[0,0], CBmonths[0,1], 1, 0, 0, 0, tzinfo=timezone('US/Eastern')) + \
    timedelta(hours=(x+0.5)))
hourlyDF_bystart["DateTime_UTC"] = hourlyDF_bystart["DateTime"].apply(lambda x: x.astimezone(pytz.utc))
```

```python
# Load weather data
dtfmt = "%Y-%m-%d %H:%M:%S"
start_weather = date(2013,7,1)
end_weather = date(2014,8,31)
weather_filename = "/home/vagrant/citibike/weatherDB_" + start_weather.strftime("%Y-%m-%d") + "_to_" \
    + end_weather.strftime("%Y-%m-%d") + ".csv"
weather_df = pd.read_csv(weather_filename, index_col=0)

# Index by proper UTC DateTimes, using nearest previous 1/2 hours (same as Citi Bike)
weather_df["DateTime_UTC"] = [datetime(*datetime.strptime(x, dtfmt).timetuple()[0:4], \
    minute=int(np.floor(datetime.strptime(x, dtfmt).minute/30.)*30), tzinfo=pytz.utc) \
    for x in weather_df["DateUTC"].values]
del weather_df["Date"], weather_df["DateUTC"], weather_df["TimeEST"], weather_df["TimeEDT"]

# Get rid of bad temperature values
weather_df["TemperatureF"].ix[weather_df["TemperatureF"] < -100] = np.nan
weather_df.head()
```

```python
weather_df[["DateTime_UTC", "TemperatureF", "PrecipitationIn", "Humidity", "Wind SpeedMPH"]].head()
```

```python
allhourly_gp = hourlyDF_bystart.groupby("DateTime_UTC")
allhourly = pd.DataFrame({"DateTime_UTC":allhourly_gp["DateTime_UTC"].first().values, \
                  "DateTimeLocal":allhourly_gp["DateTime"].first().values, \
                  "count":allhourly_gp["total count"].sum().values})
```

```python
holidays = [datetime.strptime(x, '%Y-%m-%d').date() for x in ['2013-07-04','2013-09-02','2013-10-14','2013-11-11', \
    '2013-11-28','2013-12-25','2014-01-01','2014-01-20','2014-02-17','2014-05-26','2014-07-04']]
holidayRelated = [datetime.strptime(x, '%Y-%m-%d').date() for x in ['2013-07-05','2013-11-27','2013-11-29','2013-12-24']]
```

```python
allhourly["weekday"] = allhourly["DateTimeLocal"].map(lambda x: calendar.weekday(x.year, x.month, x.day))
allhourly["hour"] = allhourly["DateTimeLocal"].map(lambda x: x.hour)
allhourly["holiday"] = allhourly["DateTimeLocal"].map(lambda x: x.date() in holidays+holidayRelated)
```

```python
allhourly = allhourly.merge(weather_df[["DateTime_UTC", "TemperatureF", "PrecipitationIn", "Humidity", "Wind SpeedMPH"]], \
                            how='left', on="DateTime_UTC")
allhourly["Wind SpeedMPH"].ix[allhourly["Wind SpeedMPH"] == '-9999.0'] = 'nan'
allhourly["Wind SpeedMPH"].ix[allhourly["Wind SpeedMPH"] == 'Calm'] = '0'
allhourly["Wind SpeedMPH"] = allhourly["Wind SpeedMPH"].map(float)
allhourly.head()
```

```python
wkday_hourly_avg = allhourly.ix[np.array(ismember(allhourly["weekday"].values, np.arange(0,5))) \
                                & ~allhourly["holiday"]].groupby("hour")["count"].mean()
wkend_hourly_avg = allhourly.ix[~np.array(ismember(allhourly["weekday"].values, np.arange(0,5)))].groupby("hour")["count"].mean()
plt.figure()
wkday_hourly_avg.plot(color='b')
wkend_hourly_avg.plot(color='r')
```

```python
tempF_avg = allhourly.groupby("TemperatureF")["count"].mean()
plt.figure()
tempF_avg.plot()
```

```python
precip_avg = pd.DataFrame({"precip":allhourly.groupby("PrecipitationIn")["PrecipitationIn"].count().index.values, \
                           "n":allhourly.groupby("PrecipitationIn")["PrecipitationIn"].count().values, \
                           "count mean":allhourly.groupby("PrecipitationIn")["count"].mean().values, \
                           "count std":allhourly.groupby("PrecipitationIn")["count"].std().values})
rain_avg = pd.DataFrame({"precip":allhourly.ix[allhourly["TemperatureF"] > 35].groupby("PrecipitationIn")["PrecipitationIn"].count().index.values, \
                           "n":allhourly.ix[allhourly["TemperatureF"] > 35].groupby("PrecipitationIn")["PrecipitationIn"].count().values, \
                           "count mean":allhourly.ix[allhourly["TemperatureF"] > 35].groupby("PrecipitationIn")["count"].mean().values, \
                           "count std":allhourly.ix[allhourly["TemperatureF"] > 35].groupby("PrecipitationIn")["count"].std().values})
snow_avg = pd.DataFrame({"precip":allhourly.ix[allhourly["TemperatureF"] < 35].groupby("PrecipitationIn")["PrecipitationIn"].count().index.values, \
                           "n":allhourly.ix[allhourly["TemperatureF"] < 35].groupby("PrecipitationIn")["PrecipitationIn"].count().values, \
                           "count mean":allhourly.ix[allhourly["TemperatureF"] < 35].groupby("PrecipitationIn")["count"].mean().values, \
                           "count std":allhourly.ix[allhourly["TemperatureF"] < 35].groupby("PrecipitationIn")["count"].std().values})
```

```python
#train on every other day
train_days = [min(allhourly.DateTimeLocal).date() + timedelta(days=x) for x in \
    range(0, (max(allhourly.DateTimeLocal).date() - min(allhourly.DateTimeLocal).date()).days, 2)]
test_days = [min(allhourly.DateTimeLocal).date() + timedelta(days=x) for x in \
    range(1, (max(allhourly.DateTimeLocal).date() - min(allhourly.DateTimeLocal).date()).days, 2)]

# Convert to boolean index vectors
wkday_train_bools = allhourly["DateTimeLocal"].map(lambda x: x.date()).isin(train_days) & allhourly["weekday"].isin(range(0,5))
wkday_test_bools = allhourly["DateTimeLocal"].map(lambda x: x.date()).isin(test_days) & allhourly["weekday"].isin(range(0,5))
wkend_train_bools = allhourly["DateTimeLocal"].map(lambda x: x.date()).isin(train_days) & allhourly["weekday"].isin(range(5,7))
wkend_test_bools = allhourly["DateTimeLocal"].map(lambda x: x.date()).isin(test_days) & allhourly["weekday"].isin(range(5,7))
```

```python
temp_array = np.array(allhourly["TemperatureF"].ix[wkday_train_bools].values).reshape(sum(wkday_train_bools),1)
wkday_train_set = np.append(temp_array, temp_array**2, axis=1)
wkday_train_set = np.append(wkday_train_set, temp_array**3, axis=1)
wkday_train_set.shape
```

```python
temp_array = np.array(allhourly["TemperatureF"].ix[wkday_train_bools].values).reshape(sum(wkday_train_bools),1)
wkday_train_set = np.append(temp_array, temp_array**2, 1)
wkday_train_set = np.append(wkday_train_set, temp_array**3, 1)
wkday_train_set = np.append(wkday_train_set, temp_array**4, 1)
wkday_train_set = np.append(wkday_train_set, \
    np.array(allhourly["PrecipitationIn"].ix[wkday_train_bools].values).reshape(sum(wkday_train_bools),1), 1)
wkday_train_set = np.append(wkday_train_set, np.array(allhourly["hour"].ix[wkday_train_bools].map( \
    lambda x: np.log(wkday_hourly_avg.ix[x])).values).reshape(sum(wkday_train_bools),1), 1)
wkday_train_set = sm.add_constant(wkday_train_set)
wkday_train_resp = np.log(np.array(allhourly["count"].ix[wkday_train_bools].values))

# Make sure no NaN's for training set
rm_nans = np.isnan(wkday_train_set).any(axis=1)
wkday_train_set = wkday_train_set[~rm_nans,:]
wkday_train_resp = wkday_train_resp[~rm_nans]

temp_array = np.array(allhourly["TemperatureF"].ix[wkday_test_bools].values).reshape(sum(wkday_test_bools),1)
wkday_test_set = np.array(allhourly["TemperatureF"].ix[wkday_test_bools].values).reshape(sum(wkday_test_bools),1)
wkday_test_set = np.append(temp_array, temp_array**2, 1)
wkday_test_set = np.append(wkday_test_set, temp_array**3, 1)
wkday_test_set = np.append(wkday_test_set, temp_array**4, 1)
wkday_test_set = np.append(wkday_test_set, \
    np.array(allhourly["PrecipitationIn"].ix[wkday_test_bools].values).reshape(sum(wkday_test_bools),1), 1)
wkday_test_set = np.append(wkday_test_set, np.array(allhourly["hour"].ix[wkday_test_bools].map( \
    lambda x: np.log(wkday_hourly_avg.ix[x])).values).reshape(sum(wkday_test_bools),1), 1)
wkday_test_set = sm.add_constant(wkday_test_set)
wkday_test_resp = np.log(np.array(allhourly["count"].ix[wkday_test_bools].values))

# Make sure no NaN's for test set
rm_nans = np.isnan(wkday_test_set).any(axis=1)
wkday_test_set = wkday_test_set[~rm_nans,:]
wkday_test_resp = wkday_test_resp[~rm_nans]
```

```python
wkday_model = sm.GLM(wkday_train_resp, wkday_train_set, family=sm.families.Poisson())
wkday_results = wkday_model.fit()
wkday_train_pred = wkday_results.predict(wkday_train_set)
wkday_test_pred = wkday_results.predict(wkday_test_set)
```
