```python
import numpy as np
import pandas as pd
import urllib2 as url
import datetime, time
import subprocess
import gzip
import matplotlib as plt
```

```python
CBmonths = np.array([np.array([np.ones((6,1))*2013, np.ones((6,1))*2014]).flatten(), \
    np.array([np.arange(7,13,1), np.arange(1,7,1)]).flatten()], dtype='int').T
mo_strs = []
for i in range(CBmonths.shape[0]):
    mo_strs.append(datetime.date(CBmonths[i,0], CBmonths[i,1], 1))
CBcsv_filename = "/home/vagrant/citibike/trip_logs/citibikeDB_" \
    + mo_strs[0].strftime("%m-%Y") + "_to_" + mo_strs[-1].strftime("%m-%Y") + ".csv.gz"
trip_df = pd.read_csv(CBcsv_filename, compression="gzip", index_col=0)
```

# Build hourly histograms of trips from and to each station and combine with weather info
# into one data frame

```python
trip_df.head()
```

```python
reftime = time.mktime(time.strptime("2000-01-01 00:00:00", "%Y-%m-%d %H:%M:%S"))
```

```python
trip_df_sub = trip_df[0:10000]
```

```python
trip_df_sub["startSecFrom2000"] = trip_df_sub.starttime.apply(lambda x: \
    time.mktime(time.strptime(x, "%Y-%m-%d %H:%M:%S"))) - reftime
```

```python
trip_df_sub["stopSecFrom2000"] = trip_df_sub.stoptime.apply(lambda x: \
    time.mktime(time.strptime(x, "%Y-%m-%d %H:%M:%S"))) - reftime
```

```python
trip_df_sub.head()
```

```python
binstart = time.mktime(time.strptime(datetime.date(CBmonths[0,0], CBmonths[0,1], 1).strftime("%Y-%m-%d %H:%M:%S"), "%Y-%m-%d %H:%M:%S"))
binend = time.mktime(time.strptime(datetime.date(CBmonths[-1,0], CBmonths[-1,1], 1).strftime("%Y-%m-%d %H:%M:%S"), "%Y-%m-%d %H:%M:%S"))
```

```python
hourbins = np.linspace(binstart-reftime, binend-reftime, (binend-binstart)/3600 + 1)
```

```python
trip_df_sub["startbin"] = np.digitize(trip_df_sub.startSecFrom2000, hourbins)
trip_df_sub["stopbin"] = np.digitize(trip_df_sub.stopSecFrom2000, hourbins)
```

```python
trip_df_sub.head()
```

```python
group_sub = trip_df_sub.groupby(["startbin", "start station id", "end station id"])
```

```python
df_sub = pd.DataFrame(group_sub["tripduration"].count())
df_sub.columns = ["total count"]
```

```python
df_sub["male count"] = group_sub["gender"].aggregate(lambda x: np.sum(x == 'm'))
df_sub["female count"] = group_sub["gender"].aggregate(lambda x: np.sum(x == 'f'))
df_sub["subscriber count"] = group_sub["usertype"].aggregate(lambda x: np.sum(x == 'Subscriber'))
df_sub["mean duration"] = group_sub["tripduration"].mean()
```

```python
trip_df_sub["birth year"] = trip_df_sub["birth year"].apply(lambda x: x.replace("\N", "0")).apply(int)
```

```python
trip_df_sub["birth year"].ix[[i-1 for i, elem in enumerate([(x < 1000) for x in trip_df_sub["birth year"]], 1) if elem]] = None
trip_df_sub["birth year"] = np.array(trip_df_sub["birth year"], dtype=float)
```

```python
df_sub["mean age"] = group_sub["birth year"].aggregate(lambda year: np.round(2014 - np.nanmean(year)))
```
